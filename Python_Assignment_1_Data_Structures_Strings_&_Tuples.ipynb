##Strings (Concatenation, Slicing, and Other Methods) :

```python

# 1. String Concatenation:

string1 = "Hello"

name = input("Enter Your Name : ")
print(f"{string1} {name}!")

string3 = "Welcome to Python Programming"

new_string = string1+" "+name+"! "+string3

print("\nOutput String :", new_string)
```

```python
# 2. String Slicing and Indexing:

print(f"First Character : {new_string[0]}")
print(f"Last Character : {new_string[-1]}")
print(f"First 5 Characters : {new_string[:5]}")
print(f"Last 11 Characters : {new_string[-11:]}")
print(f"Reverse String : {new_string[::-1]}")
print(f"Word Slicing : {new_string[26:32]}")
```

```python
# Word Slicing
# Use slicing and print the word “Python” from the existing string.

start = new_string.index("Python")
print(start)

text = new_string.index("on")
print(text)

end = len("Python")
print(end)
```

```python
# 3. String Methods:

strM = "Python beginner tutorial."
new_string = 'Hello Karunya! Welcome to Python Programming'

print(f"Uppercase : {strM.upper()}")
print(f"Lowercase : {strM.lower()}")
print(f"Captalize the Sentence : {strM.capitalize()}")

#string.count(text)

print(f"Total number of t - Occurrences : {new_string.count('t')}")

#string.replace(old_value, new_value)

strM = "Python beginner tutorial."
new_str = strM.replace("Python","Data Analytics")
print(f"Origianal Sentence : {strM}")
print(f"Replaced Sentence : {new_str}")

```
