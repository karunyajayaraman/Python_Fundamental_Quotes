#Case Study – Restricted Word Detection System
Instructions:

Create a list of words containing normal and restricted words.

Use a for loop to read words one by one.

Check whether the current word is a restricted word using if condition.

If the restricted word is found: • Display "Restricted word detected"
• Stop the loop using break

Otherwise display all safe words using print()

```python
# Case Study – Restricted Word Detection System

# List of words
words = ["Hello", "Python", "Admin", "Blocked", "Welcome"]

# Restricted word
restricted_word = "Blocked"

# for loop
for word in words:

    # Checking restricted word
    if word == restricted_word:
        print(f"Restricted word detected {word}")
        break

    # Safe words
    print("Safe words",word)# Case Study – Restricted Word Detection System
```
