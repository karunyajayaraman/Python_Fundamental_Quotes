1. List Creation:

a. Create a list named age_list with five integer elements. For eg., [24, 25, 27, 28 ,29]

```python
age_list = [24, 25, 27, 28 ,29]
name_list = ['Karunya', 'Karthikeyan', 'Nimalan', 'Reena', 'Gayathri']

print(type(age_list))
print(type(name_list))
```

#2. List Operations / Modifications:
a. Append the string "Yazhini" to name_list.

b. Insert the element 26 at index 2 in age_list.

c. Remove the string "Yazhini" from name_list.

d. Pop the last element from age_list.

e. Extend the age_list with additional ages [31,30,32].

f. Sort age_list in descending order.

g. Find the max age, min age, and sum of all ages from age_list.

```python
# a. Append() - Add the element at the end of the List

name_list.append('Yazhini')
print(f'\nAdd the Name : {name_list}')
```

```python
# b. Insert() - Insert the element at the Index 2 position

age_list.insert(2, 26)
print(f"\nAfter Insert : {age_list}")
```

```python
# c. Remove() - Remove the mentioned element from the List

name_list.remove("Yazhini")
print(f"\nAfter Removed : {name_list}")
```

```python
# d. Pop() - Remove the last added Element

age_list.pop()
print(f"\nAfter Popped : {age_list}")
```

```python
# e. Extend - Add multiple Elements to the List

age_list.extend([31,30,32])
print(f"\nAfter Extend : {age_list}")
```

```python
# d. Sort - Arrange the elements in Descending order

age_list.sort(reverse = True )
print(f'\nAfter Descending Order : {age_list}')
```

```python
# f. Max() Age & Min() Age & Sum() of Age - Returns the maximum, minimum and sum of age_list

age_list = [24, 25, 26, 26, 26, 27, 28, 31, 30, 32]

print(f'\nMaximum Age : {max(age_list)}')

print(f'\nMinimum Age : {min(age_list)}')

print(f'\nSum of Age : {sum(age_list)}')

```

#Dictionary (Creation, Modification and Access):
a. Create a dictionary named student_marks that maps the names of five students to their marks (use a scale of 0 to 100).

b. Access and print the mark of a specific student of your choice.

c. Add a new student, "Janani," with a mark of 80 to the student_marks dictionary.

d. Update the mark of any one older student to 82.

e. Use the keys(), values(), and items() methods to print all keys, values, and key-value pairs in the student_marks dictionary.

```python
student_marks = {
    "Karunya" : 85 ,
    "Karthikeyan": 90 ,
    "Gayathri": 75 ,
    "Nimalan" : 88 ,
    "Reena" : 95
    }

print(f"Students | Marks : {student_marks}")
print(type(student_marks))

print(f"\nMark of Nimalan: ", student_marks["Nimalan"])
```

```python
student_marks["Janani"] = "80"
print(f"Added Student_mark : {student_marks}")
```

```python
student_marks["Gayathri"] = 82
print(f"Updated the Student mark : {student_marks}")

print(f"\nStudent Names : {student_marks.keys()}")

print(f"\nStudent Marks : {student_marks.values()}")

print(f"\nStudent & Marks : {student_marks.items()}")
```

```python
set1 = {1, 3, 5, 7, 9}
set2 = {2, 3, 5, 8, 10}

print(f'Set 1 : {set1}')
print(f'Set 2 : {set2}')

print(f'\n Union : {set1.union(set2)}')
print(f'\n Intersection : {set1.intersection(set2)}')
```

# Operators & Conditional Statements : (IF, ELIF, ELSE)

```python
score_range = int(input("Enter the Score Range (0-10): "))

if score_range < 0 or score_range > 10:
  print(" Enter the Valid Score Range between (0-10)")

elif score_range > 7:
  print("Above Average : Excellent Work! Keep up the great Perfomance")

elif score_range >= 4 and score_range <= 7:
  print("Average: Good effort! Keep practicing, there's room for improvement.")

else: # score_range <= 4
  print("Below Average : Need to improve your performance and practice more.")
```
