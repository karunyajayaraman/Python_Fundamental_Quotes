#Case Study – Student Data Management System
Instructions:

Create a dictionary named student with the following keys: name, age, course

Store multiple students' data using lists as values.

Add one new student details to the dictionary:

• Append new name to name list

• Append new age to age list

• Append new course to course list

Display all student details using dictionary keys.

Access and display:

• First student's details using indexing

• Newly added student's details using indexing

```python
#Create a dictionary named student with the following keys: name, age, course

student={ "Name": "Nimalan",
         "Age":25,
         "Course" :"Data Engineer"
}
print(student)
print(type(student))
```

```python
#Store multiple students' data using lists as values.

student = {
    "Name": ["Karunya", "Karthik", "Gayathri"],
    "Age": [28, 24, 26],
    "Course": ["Data Science", "AI", "Machine Learning"]
}
print(student)
```

```python
#Add one new student details to the dictionary:

#• Append new name to name list
#• Append new age to age list
#• Append new course to course list

student = {
    "Name": ["Karunya", "Karthik", "Gayathri"],
    "Age": [28, 24, 26],
    "Course": ["Data Science", "AI", "Machine Learning"]
}

student["Name"].append("Nimalan")
student["Age"].append(24)
student["Course"].append("Data Engineer")

print(student)
```

```python
# Display all student details using dictionary keys.

print("All Names:", student["Name"])
print("All Ages:", student["Age"])
print("All Courses:", student["Course"])
```

```python
#Access and display:

#• First student's details using indexing

print(f"First Student Details ")
print(f"\nName : {student["Name"][0]}")
print(f"Age : {student["Age"][0]}")
print(f"Course : {student["Course"][0]}")
```

```python
#• Second student's details using indexing

print(f"Second Student Details ")
print(f"\nName : {student["Name"][1]}")
print(f"Age : {student["Age"][1]}")
print(f"Course : {student["Course"][1]}")
```

```python
#• Newly added student's details using indexing

print(f"Newly Student Details ")
print(f"\nName : {student["Name"][-1]}")
print(f"Age : {student["Age"][-1]}")
print(f"Course : {student["Course"][-1]}")
```
