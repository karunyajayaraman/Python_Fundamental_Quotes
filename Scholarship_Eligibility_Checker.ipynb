#Case Study-Scholarship Eligibility Checker
Instructions:

Create variables: student_name, marks, attendance, family_income.

Use conditional statements (if, elif, else) to:

Award Full Scholarship if marks ≥ 90 and attendance ≥ 90.
Award Half Scholarship if marks ≥ 80 or attendance ≥ 85.
Award Need-Based Scholarship if marks ≥ 70 and family_income < 30000.
Otherwise, display Not Eligible.
Display all data (no loops).

```python
# if-elif-else statements

#student Details :

student_name = input("Enter the Student Name : ")
mark = int(input("Enter your Mark : "))
attendance = int(input("Number of Attendance : "))
family_income = float(input("Family Income : "))

#Use conditional statements (if, elif, else)

if mark >= 90 and attendance >= 90:
  scholarship = "Full Scholarship"
elif mark >= 80 or attendance >= 85:
  scholarship = "Half scholarship"
elif mark >= 70 and family_income < 30000:
  scholarship = "Need-Based Scholarship"
else:
  scholarship = "Not Eligible"

print("Student Name : ",student_name)
print("Student Mark : ", mark)
print("Attendance : ", attendance)
print("Family Income : ", family_income)
print("Scholarship status : ",scholarship)
```
