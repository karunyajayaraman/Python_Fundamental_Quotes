#Case Study – Employee Attendance Count System using continue
Instructions:

Create a list named attendance with values: Present, Absent, Present, Leave, Present

Create count variables:

• present
• absent
• leave

Use a for loop to iterate through the attendance list.

Use continue statement to: • Skip remaining conditions after counting Present employees

Count: • Total Present employees
• Total Absent employees
• Total Leave employees

Display all counts using print()

```python
# Case Study – Employee Attendance Count System using continue

# Attendance list
attendance = ["Present", "Absent", "Present", "Leave", "Present"]

# Count variables
present = absent = leave = 0


# for loop
for status in attendance:

    # Count present employees
    if status == "Present":
        present += 1
        continue

    # Count absent employees
    if status == "Absent":
        absent += 1

    # Count leave employees
    elif status == "Leave":
        leave += 1

# Display counts
print("Total Present:", present)
print("Total Absent:", absent)
print("Total Leave:", leave)
```
