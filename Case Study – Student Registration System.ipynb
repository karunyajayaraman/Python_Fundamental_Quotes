# Create variables:

# student details

```python
student_name = input("Enter student name : " )
student_age = int(input("Enter student age : "))
course_name = input("Enter course name : ")
course_fee = float(input("Enter course fee : "))
```

```python
# Calculate-course_fee

gst = 0.18 * course_fee
total_fee = course_fee + gst
```

```python
# gst - decimal point- Any number with a decimal point is treated as a float
# When an int and a float are used in an arithmetic operation, Python converts the result to a float automatically.
```

```python
# Display data type

print(type(student_name))
print(type(student_age))
print(type(course_name))
print(type(course_fee))
print(type(gst))
print(type(total_fee))
```
