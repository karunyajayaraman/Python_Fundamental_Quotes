#Case Study – Student Course Registration System using Set
Instructions:

Create a set named courses with course names: Python, SQL, Power BI, Python, Excel

Perform set operations: • Add course using add()
• Add multiple courses using update()
• Remove course using remove() and discard()

Display all courses and total number of courses using len()

```python
# create set

courses ={'Python', 'SQL', 'Power BI', 'Python', 'Excel'}
print(f"Orginal Courses : {courses}")
```

```python
# Perform set operations:

# • Add course using add()
courses.add("Tableau")
print(f"After adding(): {courses}")
```

```python
# • Add multiple courses using update()

courses.update(["Machine Learning", "Data Analytics"])
print(f"After update(): {courses}")
```

```python
# • Remove course using remove()

courses.remove("SQL")
print(f"After remove('SQL'): {courses}")
```

```python
# discard()
courses.discard("Excel")
print(f"After discard('Excel'): {courses}")
```

```python
# Display all courses and total number of courses using len()

print(f"Available Courses: {courses}")
```

```python
print(f"Total Length of Courses : {len(courses)}")
```

```python
print("-----Student Course Registration System-----")

print(f"\nOrginal Courses : {courses}")

print(f"\nAfter adding(): {courses}")

print(f"\nAfter update(): {courses}")

print(f"\nAfter remove('SQL'): {courses}")

print(f"\nAfter discard('Excel'): {courses}")

print(f"\nAvailable Courses: {courses}")

print(f"\nTotal Length of Courses : {len(courses)}")
```
