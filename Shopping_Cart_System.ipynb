#Case Study – Shopping Cart System

```python
# Create a list named cart with initial items: Milk, Bread, Eggs, Apple

my_cart = ["Milk","Bread","Eggs","Apple"]
print(f"Cart List : {my_cart}")

#Perform list operations:

#• Add multiple items using extend()
#• Insert item at index 2 using insert()
#• Remove item Apple using remove()
#• Remove last item using pop()

my_cart.append("Onion")
print(f"Cart Added List : {my_cart}")

my_cart.insert(1,"Mango")
print(f"Inserted Cart List : {my_cart}")

my_cart.remove("Apple")
print(f"Removed Item from Cart List : {my_cart}")

my_cart.pop()
print(f"Removed Last Item from Cart List : {my_cart}")

#Use list methods:

#• index() to find position of Eggs
#• count() to count occurrences of Milk

print(f"Index position : {my_cart.index("Milk")}")
print(f"Index position : {my_cart.index("Mango")}")

print(f"Cart Count : {my_cart.count("Mango")}")
```

```python

#Apply indexing:

#• Display first item
#• Display last item

print(my_cart)
print(f"Display First Item : {my_cart[0]}")
print(f"Display First Item : {my_cart[-1]}")

#Apply slicing:

#• Display first 3 items from the list

print(f"Display First 3 items from the cart :{my_cart[:3]}")
```

```python
#Find total number of items using len()

print(f"Total Items : {len(my_cart)}")

#Display all outputs using print()
```
