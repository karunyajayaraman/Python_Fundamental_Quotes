# Case Study: Bank Loan Eligibility System
Instructions:

Create variables: customer_name, age, income, credit_score, existing_loan

Use conditional statements (if, elif, else) to:

• Approve loan if age ≥ 21, income ≥ 30000,

-- credit_score ≥ 700, and no existing loan

• Conditionally approve loan if credit_score ≥ 600 and income ≥ 40000

• Otherwise reject the loan application

Display customer details and loan status using print()

```python
customer_name = input("Enter the Customer Name : ")
age = int(input("Enter the Age : "))
income = float(input("Enter the Customer Income : "))
credit_score = int(input("Enter the Credit Score : "))
existing_loan = input("Existing Loan : (yes/no) : ")


# Use conditional statements (if, elif, else) to:

#• Approve loan if age ≥ 21, income ≥ 30000,

#-- credit_score ≥ 700, and no existing loan

if age >= 21 and income >= 30000:
  if credit_score >= 700 and existing_loan == "yes":
    status = "Your Loan Approved"
  elif credit_score >= 600 and income >= 40000:
    status = "Conditionally Approved"
  else:
    status = "Rejected your Loan"
else:
  status = "Not Eligible to proceed"

#Display customer details and loan status using print()

print("\nCustomer:", customer_name)
print("Age:", age)
print("Income:", income)
print("Credit Score:", credit_score)
print("Loan Status:", status)
```
