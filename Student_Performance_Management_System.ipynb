#Case Study – Student Performance Management System
Instructions

Create two lists: students, marks

Use range() to: Access and print student names and marks using index positions

Use enumerate() to: Display student index number and student name

Use zip() to: Combine student names and marks together

Display all outputs using for loops

Use proper headings with print() statements

Print the final result in a clean format

```python
students = ['Karunya', 'Karthikeyan', 'Nimalan', 'Reena', 'Gayathri']
marks = ['85','80','95','78','90']

print('==== Students Names and Marks using range() ====')

# Range()
for i in range(len(students)):
    print(f"Student: {students[i]} | Marks: {marks[i]}")

print('\n==== Student Index number and Student name using Enumerate() ====')

# Enumerate
for i, student in enumerate(students):
  print(f'Index Position : {i} | Student Name: {student}')

print('\n==== Student names and marks together using zip() ====')

# Zip

for student, mark in zip(students,marks):
  print(f'Student name : {student} | Mark : {mark}')

# Final result

print("\n=== Final Student Report ===")

for index, (student,mark) in enumerate(zip(students,marks)):
  print(f'{index}.{student:<10} - {mark} Marks')
```
