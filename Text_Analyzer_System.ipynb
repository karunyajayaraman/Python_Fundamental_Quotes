#Case Study – Text Analyzer System
Instructions:

Take a string input from the user using input().

Convert the text into lowercase using lower().

Use a for loop to iterate through each character in the string.

Count: • Total vowels • Total consonants • Total digits • Total special characters

Use: • isalpha() for alphabets • isdigit() for numbers

Display all counts using print()

```python
user_text = input("Enter the String : ").lower()  # Good Morning123

v = c = d = sc = 0

for letter in user_text:

  if letter in "aeiou":
    v += 1
  elif letter.isalpha():
    c += 1
  elif letter.isdigit():
    d += 1
  else:
    sc += 1

print(f'Vowels : {v}')
print(f'Consonants : {c}')
print(f'Digits : {d}')
print(f'Special Characters : {sc}')
```
